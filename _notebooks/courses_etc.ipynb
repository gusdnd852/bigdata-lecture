### 1. 중요도 순서 (/는 동시에, ()는 끝나고나서)
1. 선형대수(딥러닝), 확률론(통계학), ML사이언스(강화학습)
2. 보안(해킹실습->리버싱/Win32&MFC), 양자컴퓨팅/물리학, ML엔지니어링
3. 자연어처리(DL NLP), 컴퓨터비전(DL Vision), 클라우드/DevOps
4. 암호학/정수론, 웹/앱개발, 미적분학/수치해석
5. 쓰레드(가비지컬렉팅), 시스템프로그래밍, 소켓프로그래밍
6. C언어, 어셈블리어, 분산처리/Spark, 가상화
7. 그래픽스, 정보검색, 블록체인, 두뇌-컴퓨터인터페이스
<br><br>

-   통계학 : 여인권교수님(숙명여대) [통계학 1][statics1], [통계학2][statics2]듣기

-   강화학습 : [김성훈교수님(홍콩과기대)][reinforcement1] => [유튜브(팡요랩)이론][reinforcement2] + [이승철교수님(연세대)구현][reinforcement3]

-   해킹실습 : 유튜브(나동빈) [시스템해킹][hacking1], [웹해킹][hacking2], [게임해킹][hacking3], [ARP스푸핑][hacking4] 따라해보기

-   ML 엔지니어링 : [송호연님 로드맵][ml_engineering]에 있는 것들 공부 (서빙/엔지니어링 관점, JIT, Horovod 등등)
-   자연어처리 : [스탠포드대학][nlp1], [조경현교수님(뉴욕대학교)][nlp2]
-   컴퓨터비전 : [황성수교수님(한동대)][vision1], [유튜브(멈춤보단천천히라도)][vision2]
-   클라우드 : 오픈스택 강의보면서 클라우드 시스템 만들어보기
-   DevOps : 도커, 쿠버네티스, 쿠브플로우, AWS, Azure, Jenkins 등등 배워보기
-   정수론 : [유튜브(쑤튜브)][number]
-   암호학 : [조인준교수님(배재대)][cryptography1] =>[홍득조교수님(전북대)][cryptography2]

-   앱개발 : [플러터 강의][flutter], 안드로이드 Okhttp/레트로핏/ 등 유명 라이브러리 공부하기
-   미적분 : [3B1B 보기][3B1B_calculus] => 이상준교수님 [미적분학1][calculus1], [미적분학2][calculus2]

-   쓰레드 : [깃헙블로그][thread]

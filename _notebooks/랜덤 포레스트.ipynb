### 1. 랜덤 포레스트란?
랜덤 포레스트(Random Forest)는 분류, 회귀 분석 등에 사용되는 앙상블 학습 방법의 일종입니다. 무작위의의사결정 트리로 이루어진 숲이라고 생각할 수 있습니다. 랜덤 포레스트에 대한 설명에 앞서 앙상블, 부스팅, 배깅에 대한 개념을 이해하고 가겠습니다.
#### 1.1 앙상블
앙상블이란 단일 모형들 보다 더 나은 예측 모형을 얻기 위해 **여러 모형**을 사용하는 기법입니다.
#### 1.2 부스팅(boosting)
부스팅이란 부트스트랩 표본을 이용하여 여러 개의 분류기를 얻고 통합된 분류기를 얻는 방법입니다. 이전 분류기에 의존하여 다음 분류기 생성시 가중치를 더하여 선택될 확률을 높여 순차 생성합니다.
#### 1.3 배깅(bagging)
배깅이란 부트스트랩 표본을 이용하여 여러 개의 분류기를 얻고 하나의 분류기를 생성하는 방법입니다. 이전 분류기와 상관없이 독립적으로 생성합니다.
#### 1.4 랜덤 포레스트(Random Forest)
랜덤 포레스트는 여러 개의 의사결정트리를 만들고 각각의 예측 값들을 조합하여 정밀도가 높은 분류를 하는 기법입니다. 배깅을 수정한 방법을 이용하고 랜덤으로 변수를 선택하는 방법을 이용한다. 그리고 랜덤성에 의해 트리간의 상관관계 감소시켜 분산의 감소를 향상시킨다.

랜덤 포레스트의 알고리즘은 아래와 같다.

1) 복원추출방법을 이용하여 training data로부터 N개의 부트스트랩 샘플을 추출한다.

2) 전체 설명변수 중 랜덤하게 m개의 변수를 선택하고 m개의 변수들 중에서 최적의 변수와 분리점을 선택한다.

3) 2)과정을 각각의 끝 노드에서 최소 노드 크기에 도달할 때까지 반복한다.
 
=> 1) 2) 3) 과정을 B번 반복한 후, vote/average을 구하여 예측


### 2. 랜덤 포레스트의 파라미터
#### 2.1포레스트의 크기(트리의 개수) $T$
총 포레스트는 몇개의 트리를 구성할 지를 결정하는 파라미터이다. $T$를 작게하면 일반화 능력이 떨어져 임의의 데이터에 대해 틀린 결과를 낼 가능성이 높아진다. 반면에 $T$의 크기가 크다면 훈련과 테스트 시간이 커진다. 하지만 작은 포레스트보다 결과값이 연속적이며 일반화 능력이 우수하다.
#### 2.2 최대 허용 깊이 $D$
하나의 트리에서 루트 노드부터 중단 노드까지 최대 몇 개의 노드를 거칠 것인지를 결정하는 파라미터이다.최대 허용 깊이가 작으면 underfitting이 일어나고 크면 overfitting이 일어나기 때문에 적절한 값을 설정하는 것이 중요하다.

### 3. 랜덤 포레스트의 장점
- 랜덤 포레스트의 장점은 의사결정트리를 어떻게 만드는 지만 알고 있다면 이들의 결괄르 단순히 voting등을 종합하기만 하면 되기 때문에 굉장히 쉽게 이해하고 구현할 수 있다.
- 의사결정트리와 다르게 Overfitting이 잘 되지 않는다. 새로운 데이터에 대해 굉장이 잘 일반화되는 편이다.
- 학습이 빠르다.

### 4. 랜덤 포레스트의 단점
- 메모리 사용랑이 굉장히 많다.
- 데이터의 양이 증가해도 눈에 띄는 성능 향상이 일어나지 않는다.

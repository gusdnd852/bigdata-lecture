### 1) 인공지능이란?

![image](https://user-images.githubusercontent.com/38183241/87246935-96ef8800-c48b-11ea-9819-ea324a555b8a.png)

흔히들 인공지능 하면 영화 터미네이터 또는 알파고 등이 머릿속에 가장 먼저 떠오르시죠? 인공지능(Artificial Intelligence)이라는 단어는 힘(F=ma)이나 거리(S=vt)처럼 정확하게 정의된 용어가 아닙니다. 때문에 인공지능의 정의는 여러가지가 있는데, 여러분들에게 가장 와닿게 쉽게 설명하자면 **인공지능이란 인간처럼 어떤 것을 판단하는 능력을 실현한 기술**을 통칭해서 말합니다. 

즉, 인간처럼 스스로 생각하고 행동할 수 있는 것(터미네이터)도 인공지능이라고 부를 수 있고, 인간이 해야할 일들을 기계가 대신해서 수행할 수 있으면(알파고) 그 것도 인공지능이라고 부를 수 있습니다. 이렇게 인공지능은 크게 두가지 종류로 나뉩니다. 아래 챕터에서 좀 더 알아볼까요?


### 2) 인공지능의 종류

- 강 인공지능 : 터미네이터와 같이 마음을 가지고 스스로 생각하는 인공지능
- 약 인공지능 : 알파고, 얼굴인식, 음성인식과 같이 한 분야만 잘하도록 만들어진 인공지능

![](https://post-phinf.pstatic.net/MjAxOTAzMDdfMTcx/MDAxNTUxOTE5NDE1MjU2.6Qirmc0LDiBsga0Ww1PCRHyCTM9cstG3zfFV7rtrTjIg.BYK3C_8FWwsPTF4voezbZ9D5dsrHHHvSi15Ms_Ko8PYg.JPEG/50.jpg?type=w800_q75)

일반적으로 말하는 인공지능은 두가지 종류가 있습니다. 첫번째로 강한 인공지능은 터미네이터나 아이언맨에 나오는 비전, 울트론, 영화 AI에 나오는 데이비드와 같이 스스로 마음을 가지고 사고하고 행동할 수 있는 지능체를 말하는데 아직까지 강 인공지능은 개발되지 않았습니다. 

![](https://news.unist.ac.kr/kor/wp-content/uploads/2017/05/%EC%9D%B4%EC%8A%B9%EC%B2%A0-%EA%B5%90%EC%88%98%ED%8C%80_main-800x448.jpg)

그에 비해 약한 인공지능은 제한적인 영역에서 어떤 것을 판단하여 사람의 일을 대신 수행할 수 있는 인공지능을 말합니다. 우리가 이미 잘 알고있는 알파고와 같은 인공지능들이 이에 해당합니다. 그러면 우리는 이러한 인공지능을 어떻게 구현할 수 있을까요? 인공지능을 구현하는 방법 역시 두가지가 있습니다. 하나는 Top Down이고 나머지 하나는 Bottom Up 방식입니다. 이에 대해 자세히 알아봅시다.



### 3) 인공지능을 구현하는 두가지 방법론
- Top Down : 이미 완성된 지능체의 지식이나 생체정보를 바탕으로 인공지능 개발
- Botton Up : 수학/컴퓨터 프로그래밍 등으로 처음부터 설계하여 인공지능 개발

#### 3-1) Top-Down 개발방식

인공지능을 개발하는 방법론도 역시 크게 두가지로 나눌 수 있습니다. 바로 Top-Down과 Botton-UP입니다. **Top-Down의 경우, 실제 지성체가 가진 지성으로부터 인공지능을 개발하려는 방식**입니다. 
미국에서 진행중인 [Openworm프로젝트](https://www.youtube.com/watch?v=iVKtELhFNU4)에서는 이미 예쁜 꼬마선충이라는 벌레의 신경구조를 컴퓨터 시뮬레이션이나 실제 로봇에 삽입하여 프로그래밍된 소스코드 없이 자의적으로 벽을 만나면 피해서 움직이는 로봇을 만들어 냈습니다. 그러나 아직은 이러한 벌레 수준의 신경구조만 파악 가능하여 매우 미약한 수준이고, 사람과 같은 고등생물의 복잡한 신경구조를 연구하기에는 기술적으로 갈 길이 먼 상황입니다.

![img](https://post-phinf.pstatic.net/MjAxOTAzMjJfODIg/MDAxNTUzMjIzNTI1NzQz.qNXdCD6kXBHDwZpaRz3GiWbzYK6-sXs-jbHamvj0PfUg.e2D4kySLINOLNQjgF_iRz8kSgnPysNy1VgHA8BSkSWkg.PNG/mug_obj_2019032211584639.png?type=w1080)

한편, 1980년대에는 인공지능을 개발하기 위해 사람(전문가)이 가지고 있는 방대한 양의 지식을 다양한 규칙과 추론사슬이라는 형태로 만들어서 전문가 시스템이라는 것을 만들었고 사용자에게 결과를 제시하려고 했습니다. 이를 위해 방대한 양의 정보를 규칙으로 표현하고 그 규칙 하나하나를 컴퓨터에 프로그래밍하였으나, 저장해야하는 정보가 너무나 방대하며, 규칙화 할 수 없는 정보들도 많았기 때문에 이러한 방식으로 사람과 같은 인공지능을 완벽하게 구현하는 것은 불가능에 가까웠습니다. 또한 대량으로 입력해둔 정보들에 대한 유효성을 제대로 입증할 수도 없었기 때문에 결국 많은 사람들은 전문가 시스템을 외면했습니다.

![](https://user-images.githubusercontent.com/38183241/87243846-4caedc80-c474-11ea-99a1-5353ad33c5d4.png)

그에 비해 Bottom Up 구현 방식은 실제로 지성체의 정보나 지식 없이, 컴퓨터가 이 세상의 다양한 데이터를 스스로 보고 학습하는 방식으로 인공지능을 구현해보려는 시도입니다. 대표적인 Bottom Up 구현 방식으로는 우리가 앞으로 배울 머신러닝이 있습니다. 다음 챕터에서 더 자세히 알아봅시다.

#### 3-2) Bottom Up 구현방식

전문가시스템은 모든 규칙을 컴퓨터에게 가르치고, 그 것을 if-else 구조를 통해 컴퓨터에 구현하여 다양한 문제를 해결하려고 했습니다. 이러한 방식의 프로그래밍 방식을 규칙기반(Rule-base) 방식이라고도 라고 하며, 요즘날에도 로그인시스템, 회원가입시스템, 은행결제시스템 등등 이 세상에 있는 거의 대부분 컴퓨터 기반 시스템은 이 규칙 기반으로 구현되어있습니다. 
<br><br>

![](https://user-images.githubusercontent.com/38183241/87244900-c7c7c100-c47b-11ea-92f3-cfb0ce907718.png)

정말 넓게 보자면 로그인 시스템도 하나의 인공지능이라고 할 수 있습니다. 아이디와 비밀번호를 확인하여 이 사람이 회원인지 아닌지를 사람을 대신해 판단하기 때문입니다. 로그인 시스템의 경우 아이디와 비밀번호를 입력 받고, 이 아이디와 비밀번호가 데이터베이스에 저장되어 있는 경우, 저장되어 있지 않은 경우로 규칙을 구분할 수 있습니다. 즉, 이렇게 우리가 발생할 수 있는 상황에 대한 모든 규칙을 완벽하게 알고 있다면 규칙기반으로 시스템을 완벽하게 프로그래밍 해낼 수 있습니다. **그러나 세상에는 사람이 그 규칙을 100% 설명할 수 없는 경우나, 규칙의 갯수가 너무 많아서 모든 경우를 프로그래밍 하는 것이 어려운 경우도 분명 존재**합니다. 그런 경우는 어떻게 해야할까요?

<br><br>

![img](https://www.viva100.com/mnt/images/file/2016y/02m/19d/20160217010004448_1.jpeg)

예를 들어 사진을 하나 입력받아서 고양이와 강아지의 사진을 구분하는 규칙을 만든다고 가정해봅시다. 귀가 내려가고 인상이 부드러우면 강아지이고, 귀가 뾰족하고 눈매가 매서우면 고양이라고 100% 확신할 수 없습니다. 종에 따라서 귀가 뾰족한 강아지도 있고 눈매가 부드러운 고양이도 있습니다. 이런 경우, 어떻게 사진을 분류하는 규칙을 정할 수 있을까요? 차라리 방대한 양의 사진을 컴퓨터에게 보게 해서 스스로 규칙을 찾아내게 하는 편이 낫지 않을까요?

<br><br>

![image](https://user-images.githubusercontent.com/38183241/87245661-2728cf80-c482-11ea-89fd-963fdf41c4a1.png)

또 사람의 말을 예시로 들어봅시다. 위와 같이 사람의 말은 맥락과 분위기에 따라 달라질 수 있고, 그 경우의 수는 사람이 헤아리기 힘든 수준입니다. 이 규칙을 전부 다 찾아내서 컴퓨터에 프로그래밍 하는 것이 가능할까요? 마찬가지로 차라리 방대한 양의 글을 컴퓨터에게 읽게 해서 스스로 규칙을 찾아내게 하는 편이 낫지 않을까요? 바로 이럴 때 우리는 머신러닝을 사용합니다. 

<br><br>

![](https://user-images.githubusercontent.com/38183241/87244749-8e428600-c47a-11ea-8921-fbe87ffaf72f.png)

**머신러닝은 해당 시스템의 규칙이 정확히 무엇인지 모르는 문제에 적용**합니다. 규칙을 입력하지 않는 대신 입-출력 데이터를 많이 주고 컴퓨터에게 그 규칙을 찾게끔 하는 개발 방식이 바로 머신러닝 방법입니다. 위의 그림처럼 입력과 규칙을 주고 결과를 찾게하는 것이 일반적인 컴퓨터 시스템을 구현할 때 주로 사용하는 규칙기반(Rule-base)방식이고, 입력과 출력을 가지고 있는데, 규칙을 모를 때 사용하는 것이 머신러닝(Machine Learning) 방식입니다. 

<br><br>

![image](https://user-images.githubusercontent.com/38183241/87245150-11b1a680-c47e-11ea-8eba-002b2da49cb4.png)

기본적으로 인간은 눈이나 귀와 같은 신체기관을 통해 다양한 입력을 받고, 말과 행동을 통해 그에 대한 출력을 하면서 세상의 규칙을 배워나갑니다. 예를 들자면, 아이가 말을 잘 들을 때마다 사탕을 주면, 아이는 "말을 잘 들으면 사탕을 받는구나"라는 규칙을 터득하고, 사탕을 받기 위해 이전보다 말을 잘 듣게 됩니다. 이러한 방식을 기계가 흉내내도록 시도하는 방식이 머신러닝입니다. 
<br><br>

그래서 우리는 인간의 지능을 구현하려는 인공지능 역시도 머신러닝으로 구현하려고 시도하는 것입니다. 지능을 만들어내는 방법은 잘 모르지만, 그 결과는 너무나 잘 알고 있기 때문입니다. 그렇다면 요즘 자주 들리는 단어인 딥러닝은 머신러닝과 어떻게 다른 것일까요? 이에 대해 다음 챕터에서 자세히 알아봅시다.

### 4) 딥러닝과 머신러닝
<br>

![image](https://user-images.githubusercontent.com/38183241/83330633-6680cf80-a2cb-11ea-8e13-ece9bf71d7eb.png)

- 머신러닝 : 데이터를 이용해 규칙을 찾아내는 프로그래밍 방법
- 인공 신경망 : 머신러닝 기법 중 한가지, 인간의 뇌를 본 뜬 신경망 구조를 사용한 것
- 딥러닝 : 인공신경망의 세부분류, 인공 신경망의 신경망을 깊게 연결해서 사용하는 방법

<br><br>

머신러닝은 데이터를 이용해 규칙을 찾아내는 프로그래밍 방법이라고 앞서 이야기했습니다. 그러나 아무 데이터나 주고 맘대로 찾아보라고 할 수는 없고 기본적으로 어떤 방법으로 규칙을 찾아낼 것인지는 컴퓨터에게 알려줘야합니다. 
<br><br>

![image](https://user-images.githubusercontent.com/38183241/87245936-a1f2ea00-c484-11ea-8630-c7f131316516.png)

예를 들면 집의 크기와 집의 가격이 적힌 데이터셋이 있다고 가정해봅시다. 우리는 집의 크기가 커지면 가격도 높아질 것이라는 것을 쉽게 생각할 수 있으나, 크기가 얼마나 커지면 가격이 얼마나 높아질 것인지에 대한 디테일한 규칙은 모릅니다. 이런 경우에 우리는 컴퓨터에게 "크기가 얼마나 클때 얼마나 비싼지는 모르지만... 아마 집의 크기와 가격은 비례관계일거니까 너가 선형방정식 형태로 찾아봐"라고 알려줘야합니다. 

<br><br>

![image](https://user-images.githubusercontent.com/38183241/87246000-27769a00-c485-11ea-9035-60af2197ced5.png)

또 다른 예시를 보자면, 만약 메일의 제목과 내용, 그리고 스팸 여부가 적힌 데이터 셋이 있다고 가정해봅시다. 우리는 특정 단어(e.g. 홍보, 도박, 유흥 등)이 들어가 있으면 이 메일이 스팸메일이 높다는 사실을 알고 있습니다. 때문에 우리는 컴퓨터에게 "음.. 어떤 단어가 들어갈 때 스팸일 확률이 높은지는 모르지만 너가 단어마다 스팸일 확률 계산해보고, 그런 단어가 많이 들어갈수록 스팸메일로 분류하는 확률방식을 써서 찾아봐"라고 알려줘야합니다. **이렇게 어떻게 규칙을 찾아내는지에 대한 방법들을 바로 모델이라고 합니다.**

<br><br>

![img](https://miro.medium.com/max/2332/1*xMwqdbPMm3l_1cOVHAUYcQ.png)

위와 같이 머신러닝 모델은 매우 다양합니다. 아마 전체 다 세면 100개도 넘을 것입니다. 나중에 자세히 배우겠지만, 이 많은 방법 중에서 인간의 신경망을 한번 흉내내보자는 인공신경망 방식이 있습니다. 인공신경망 방식은 매우 옛날에 제안된 아이디어지만 현실적인 한계에 의해 제 성능을 못보여주는 계륵과 같은 존재였습니다. 그러나 그러한 문제가 해결되고 요즘날에는 인공신경망은 흔히 Universal Function Approximator라고 불리는데, 일반적으로 패턴이 보이는 데이터라면 어떤 종류든 상관 없이 규칙(함수)을 찾아낼 수 있는 궁극의 머신러닝 모델이라는 뜻입니다. 실제로 인공신경망이 규칙을 제일 잘 찾아내는 편입니다.

<br><br>

![image](https://user-images.githubusercontent.com/38183241/87246219-e1bad100-c486-11ea-84d5-07665b2259cf.png)

인공 신경망을 컴퓨터로 구현할 때 신경망 층의 개수를 몇 층으로 할지 개발자가 정할 수 있는데, 층을 너무 얇게하면 성능이 별로 좋지 않고, 층을 너무 많게 하면 데이터와 컴퓨팅 리소스가 극도로 많이 필요해져서 1980~2000년대 초반까지는 SVM등의 다른 가벼운 머신러닝 모델들이 유행했습니다. 이 때, 층이 깊은 인공신경망 구조를 써서 머신러닝을 수행하는 것을 딥러닝이라고 합니다. 즉, 딥러닝은 많고 많은 머신러닝 구현 방식 중 한가지입니다. 몇 층 이상의 신경망이 딥러닝이라고 딱히 규정하진 않습니다만, 보통 3 ~ 5층 이상 쌓으면 그 때부터 딥러닝이라고 부르는데 요즘은 100층, 1000층으로 설계된 모델들도 있습니다<br><br>

현실적인 한계(컴퓨팅 리소스, 데이터 부족)에 의해 딥러닝 연구는 계속해서 실패로 돌아갔고, 논문 제목에 Neural Network만 적혀있으면 Reject되던 딥러닝의 암흑기가 지나고 2012년이 되었습니다. 캐나다 토론토 대학의 제프리 힌튼 교수님이 이끄는 연구팀이 AlexNet이라는 이름을 가진 딥러닝 기반 이미지 분류 모델을 발표하였는데, 이 모델이 기존 이미지 분류 모델들의 성능을 크게 상회하고 세계대회를 우승하면서 딥러닝은 다시 화려하게 부활했습니다.

<br><br> 

![](https://img1.daumcdn.net/thumb/R800x0/?scode=mtistory2&fname=https%3A%2F%2Ft1.daumcdn.net%2Fcfile%2Ftistory%2F99FEB93C5C80B5192E)
<br><br> 

AlextNet은 8층의 깊은 신경망구조를 사용하였는데, 대회용 이미지 데이터셋을 더욱 조각내고 회전시켜서 이미지 데이터의 수를 늘리고, 늘어난 컴퓨팅 리소스는 Nvidia GPU를 연산에 사용해서 해결하는 등 기존 딥러닝이 가진 단점들을 해결하였더니 바로 엄청난 성능을 보여주었습니다. 이런 아이디어들 말고도 기존 딥러닝의 문제를 해결하는 매우 많은 아이디어들이 AlexNet 논문에서 제안되었고, 그 방법들은 대부분은 지금도 쓰입니다. <br><br>

![img](https://www.researchgate.net/publication/331540139/figure/fig3/AS:733273504378885@1551837435876/The-performance-of-deep-learning-with-respect-to-the-amount-of-data.ppm)

2012년 이후로 AlexNet연구팀이 제안한 방법들을 활용하였더니 딥러닝 모델들의 성능이 다른 모델들의 성능을 크게 압도하게 되면서 요즘은 머신러닝 연구라고 하면 거의 딥러닝 연구가 90% 이상을 이루게 되었습니다. 요즘에는 딥러닝이 사실상 머신러닝 연구의 전부라고 말해도 무방합니다<br><br>

자 이렇게 인공지능부터 시작해서 두 종류의 인공지능과 인공지능을 개발하는 많은 방법들에 대해 알아봤습니다. 규칙기반의 프로그래밍은 이미 프로그래밍 언어시간에 충분히 배우고 있기 때문에 이 강의에서는 여러가지 방법의 머신러닝 기법과 그리고 머신러닝 기법중 성능이 제일 좋은 딥러닝에 대해 주로 강의할 예정입니다. 그렇다면 이러한 머신러닝 기술은 어디에 쓰일 수 있을까요?

### 5) 머신러닝의 활용처


사실 머신러닝 기술은 우리의 생활 곳곳에 이미 수많이 들어가 있습니다. 예를 들면 사진을 찍으면 눈을 크게, 턱을 가름하게 만들어주는 앱을 보신적 있으실겁니다. 페이스북에 사진을 올리면 얼굴 부분을 잡아내서 친구를 태그할 수 있습니다. 쿠팡이나 11번가에 들어가면 내가 좋아할만한 물건들만 추천해주며, 유튜브는 우리가 좋아할만한 영상을 추천합니다. 구글 번역기는 내가 한국어로 문장을 적으면 영어로 번역해주고, 인공지능 스피커는 TV를 끄라는 음성을 알아듣고 TV를 끕니다. 

<br><br>

![image](https://user-images.githubusercontent.com/38183241/87246673-e2a13200-c489-11ea-976d-78c02de79d71.png)

<br><br>

이런 종류의 프로그램을 어떻게 구현할까요? if-else 등의 규칙으로 사진에서 사람의 얼굴이나 턱과 눈의 위치를 찾아내고 각 유저마다 좋아하는 물건이나 영상을 알아낼 수 있으며, 문장을 분석해서 사람이 무슨말을 했는지 알아채서 다른 언어로 번역하고, 음성 신호를 분석해서 TV를 끄라는 의도를 알아챌 수 있나요? 이미 우리의 삶 여기저기에 머신러닝 기술들이 굉장히 폭넓게 쓰이고 있습니다. 물론 현존하는 대다수의 컴퓨터 시스템은 규칙기반 방식으로 구현되어있기 때문에 규칙기반 프로그래밍 역시 굉장히 중요합니다만, 규칙기반으로는 이렇게 규칙을 우리가 명확히 알기 어려운 시스템은 구현하기 어렵습니다. (가능하다고 해도 성능이 좋지 않습니다)<br><br>


### 6) 데이터 마이닝과 데이터 사이언스

이 뿐만 아닙니다. 머신러닝은 데이터마이닝에도 매우 효과적입니다. 공장의 제품 불량율 데이터를 머신러닝으로 분석하였더니 공정을 어떤식으로 바꿔야 생산량이 제일 많아진다거나, 마트 영수증 데이터를 분석해봤더니 20대 신혼부부가 기저귀를 사면서 맥주를 같이 산다는 사실을 파악하여 기저귀 코너 옆에 맥주코너를 배치하여 수익을 올리는 등으로도 활용이 가능합니다. 즉, 꼭 어떠한 소프트웨어의 형태로 제작되어야 하는 것이 아니라 세상의 많은 데이터를 분석하는하여 이익을 창출하는 용도로도 머신러닝 기법이 굉장히 많이 활용됩니다. <br><br>

![image](https://user-images.githubusercontent.com/38183241/87247226-66105280-c48d-11ea-8701-ee6c1ec6348e.png)

<br><br>

앞선 두 예시(공장, 영수증)처럼 우리가 잘 모르고 있던 정보를 데이터에서 찾아내는 것을 바로 데이터마이닝(데이터를 캐낸다고 합니다, Mining)이라고 합니다. 인공지능과는 약간 다른 영역이지만, 머신러닝을 충분히 활용할 수 있는 영역입니다. 빅데이터 프로젝트에도 매우 유용하겠죠? 

<br><br>
![image](https://user-images.githubusercontent.com/38183241/83331448-73082680-a2d1-11ea-971f-57b568f53a86.png)

또한 이러한 빅데이터를 분산시켜서 저장하거나 처리하는 등의 데이터베이스, 빅데이터 프로세싱 기술도 각광받고 있고, 이렇게 데이터를 이용하여 컴퓨터로 뭔가를 해보려는 것을 전부 묶어서 데이터사이언스라고 합니다. 여러분은 이번 방학동안 데이터 사이언스에 대해 배우고 있는 것입니다. 물론 데이터 사이언스도 역시 컴퓨터 사이언스라는 거대한 영역의 일부입니다. 자, 이제 우리가 배울 인공지능과 머신러닝이라는 용어에 대해 잘 알게 되었습니다. 이제 머신러닝의 세부 분류에 대해 배워보겠습니다.
